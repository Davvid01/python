```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns

x=np.linspace(-5,5,100) #generuje wektor
y=x**2
print(x)
plt.plot(x,y,color='k') #łączy x i y linią
plt.xlabel('x')
plt.ylabel('y')
plt.title('Wykres funkcji kwadratowej y=x^2')
```
