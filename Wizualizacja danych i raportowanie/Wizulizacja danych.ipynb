```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import scipy.stats as stats
import seaborn as sns

x=np.linspace(-5,5,100) #generuje wektor
y=x**2
print(x)
plt.plot(x,y,color='k') #łączy x i y linią
plt.xlabel('x')
plt.ylabel('y')
plt.title('Wykres funkcji kwadratowej y=x^2')
```

```python
#zadanie 10
x=np.linspace(-15,20,100)
plt.plot(x,stats.norm.pdf(x,loc=0,scale=1),color='r',label='N(0,1)')
plt.plot(x,stats.norm.pdf(x,loc=0,scale=3),color='b',label='N(0,1)')
plt.plot(x,stats.norm.pdf(x,loc=0,scale=5),color='g',label='N(0,1)')
plt.legend(loc='upper right')
```

```python
#zadanie12
mu = 0 
sigma = 1
x = np.linspace(mu-4*sigma, mu+4*sigma,100)
plt.plot(x,stats.norm.pdf(x,loc=mu,scale=sigma), color='r')
plt.axvline(x=mu, color='k')
plt.axvline(x=mu-3*sigma, color='k')
plt.axvline(x=mu-3*sigma, color='k')
plt.axvline(x=mu+3*sigma, color='k')
plt.fill_between(x,stats.norm.pdf(x,loc=mu,scale=sigma),
                 where=(x>=mu-3*sigma)) #obcina kolor poza mu -3 na lewo
                      
```
